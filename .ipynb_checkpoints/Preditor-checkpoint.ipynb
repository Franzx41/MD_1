### Lidando com dados categoricos
##### Removemos a coluna hero_names e usamos one-hot-encoding nas features categoricas exceto por Gender, que substituimos manualmente os valores de male para 0 e female para 1

```python
data = pd.read_csv('data/clean_data.csv')
data.head()
```

```python
data.shape
```

```python
def preprocessing(data):
    data.drop("hero_names", axis=1, inplace=True)
    data = data.replace(['Male', 'Female'], [0,1])
    data = pd.get_dummies(data)
    return data
```

```python
data = preprocessing(data)
```

```python
data.shape
```

```python
types = set(data.dtypes)
types
```

```python
data_alignment = pd.read_csv('data/clean_data_for_alignment.csv')
data_alignment.head()
```

```python
data_alignment.shape
```

```python
def preprocessing_alignment(data):
    data.drop("hero_names", axis=1, inplace=True)
    data = data.replace(['Male', 'Female'], [0,1])
    label = data['Alignment']
    data = data.drop('Alignment', axis=1)
    data = pd.get_dummies(data)
    data = pd.concat((data, label), axis=1)
    return data
```

```python
data_alignment = preprocessing_alignment(data_alignment)
data_alignment.shape
```

### Predies

```python
def evaluation(data, label, most_correlated=0, average='binary'):
    model = Model(data, label, most_correlated)
    print("RESULTADO:")
    model.train_test(average=average)
    print("\nProcurando melhores parametros para a arvore de deciso...\n")
    grid = model.find_best_parameters(average=average)
    print("RESULTADO USANDO MELHORES NA ARVORE DE DECISO:")
    model.set_best_parameters(grid.best_params_)
    model.train_test(average=average)
    
```

```python
evaluation(data, "Flight")
```

```python
evaluation(data, "Super Strength")
```

```python
evaluation(data, "Accelerated Healing")
```

```python
evaluation(data_alignment, "Alignment", average='macro')
```

## Usando mais correlacionados

```python
evaluation(data, 'Flight', most_correlated=150)
```
