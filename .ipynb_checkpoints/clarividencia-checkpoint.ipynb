### Tipos de representações para dados ausentes:
- NaN
- -99 para números
- '-' para categoricas

```python
# Todas as colunas com poderes tem pelo menos um valor true
num_powers = poderes.select_dtypes(include=bool).sum(axis = 0)
num_powers[num_powers == 0]
```

```python
# Todas herois nesse dataset tem pelo menos um poder
num_powers = poderes.select_dtypes(include=bool).sum(axis = 1)
num_powers[num_powers == 0]
```
