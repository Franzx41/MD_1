```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 200)
```

```python
data = pd.read_csv("data/merged_data.csv")
data.head()
```

```python
from sklearn.metrics import matthews_corrcoef
```

```python
binary_data  = data.select_dtypes(include='bool')
poderes_corr = binary_data.corr(method=matthews_corrcoef)
```

#### A ideia de fazer um heatmap com correlaes no foi muito boa, tem muita variavel. Porm esse resultado pode ser usado na hora de fazer o modelo, podemos selecionar por exemplo os 20 poderes mais corelacionados e usar para treinar ao invez de tudo

```python
ax = sns.countplot(x="Flight", data=data)
total = data['Flight'].count()
ax.set(ylabel='Quantidade')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format((height/total)*100),
            ha="center")
plt.show()
```
