# Integração dos dataset de características de herois e superpoderes:

```python
data = herois_nan.merge(poderes, how='right', on=['hero_names'])
data.head()
```

```python
data.shape
```

```python
# não tem nenhum dado com todas os valores das colunas iguais
# porem tem alguns com mesmo nome já que o numero de linhas é 
# maior que o numero de linhas na dataset de poderes

data.drop_duplicates().shape
```

```python
data.to_csv('data/merged_data.csv',index=False)
```

### Calculando o ranking usando apenas o dataset de poderes

```python
features = ["Agility", "Accelerated Healing", "Lantern Power Ring", "Dimensional Awareness"]
hero_name = 'Batman'

jaccard = ranking(poderes.copy(), features, hero_name, metodo='jaccard')
```

```python
jaccard.head()
```

```python
test = poderes.loc[poderes["hero_names"].isin(jaccard.head()["hero_names"])]
test[features]
```

```python
features = ["Agility", "Accelerated Healing", "Lantern Power Ring", "Dimensional Awareness"]
hero_name = 'Batman'
hamming = ranking(poderes.copy(), features, hero_name, metodo='hamming')
```

```python
hamming.head()
```

```python
test = hamming.loc[hamming["hero_names"].isin(jaccard.head()["hero_names"])]
test
```
